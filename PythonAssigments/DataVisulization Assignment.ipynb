In this assignment students have to transform iris data into 3 dimensions
and plot a 3d chart with transformed dimensions and colour each data
point with specific class.

# Answer

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
import pandas as pd
```

```python
#importing dataset
iris_dataset = datasets.load_iris()
```

```python
iris_dataset
```

```python
X = iris_dataset.data
y = iris_dataset.target
```

```python
X.shape
```

```python
y.shape
```

```python
type(X)
```

```python
type(y)
```

```python
#transforming to 3d array using pca

pca = decomposition.PCA(n_components=3)
pca.fit(X)
X_transformed = pca.transform(X)
```

```python
X_transformed.shape
```

```python
#from the description we know that it has 3 classes.

class_labels = ['setosa', 'versicolor', 'virginica']
```

```python
y.shape[0]
```

```python
class_1=[]
class_2=[]
class_3=[]
for i,j in enumerate(range(X_transformed.shape[0])):
    
    if y[j]==0:
        class_1.append(X_transformed[i])
    if y[j]==1:
        class_2.append(X_transformed[i])
    if y[j]==2:
        class_3.append(X_transformed[i])
    
```

```python
print('The lenght of each class 1 is',len(class_1))
print('The lenght of each class 2 is',len(class_2))
print('The lenght of each class 3 is',len(class_3))
```

```python
np.mean(class_1)
```

```python
#Plotting the 3d graph

fig = plt.figure(1, figsize=(10, 10))
ax = plt.axes(projection ='3d')
for k in range(len(class_1)):
    ax.scatter(class_1[k][0], class_1[k][1], class_1[k][2],cmap=plt.cm.nipy_spectral,color ='green')
    
ax.text3D(-3,-1,-0.6,class_labels[0] ,horizontalalignment='center',fontsize='xx-large',
          bbox=dict(alpha=.5, edgecolor='w', facecolor='w'))
    
for l in range(len(class_2)):
    ax.scatter(class_2[l][0], class_2[l][1], class_2[l][2],cmap=plt.cm.nipy_spectral,color ='red')
    
ax.text3D(-1,0,-0.6,class_labels[1] ,horizontalalignment='center',fontsize='xx-large',
          bbox=dict(alpha=.5, edgecolor='w', facecolor='w'))
    
for m in range(len(class_1)):
    ax.scatter(class_3[m][0], class_3[m][1], class_3[m][2],cmap=plt.cm.nipy_spectral,color ='yellow')
    
ax.text3D(3,0.5,-0.6,class_labels[2] ,horizontalalignment='center',fontsize='xx-large',
          bbox=dict(alpha=.5, edgecolor='w', facecolor='w'))

```
